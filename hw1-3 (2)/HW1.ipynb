```python
# import libraries and load the csv using the source provided
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("auto-mpg.csv")
df.head()
```

```python
corr_df = df.drop(columns=["origin"])
corr_matrix = corr_df.corr(method="pearson")

plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="inferno",
    vmin=-1,
    vmax=1,
    center=0
)
plt.title("Pearson Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df, hue="origin", diag_kind="hist")
plt.show()
```

The attribute I would least likely chose is acceleration. As seen in the Pearson Correlation chart, acceleration holds a value closest to 0, which means that it has the weakest linear relationship with mpg. This means that acceleration would be the least useful predictor.
